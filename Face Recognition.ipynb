# Installing OpenCV

## Ensure latest updates are applied
Update the OS<br>
`sudo apt-get update && sudo apt-get upgrade`

## Install prereqs
`sudo apt-get install build-essential cmake unzip pkg-config`

## Install additional packages
`sudo apt-get install libjpeg-dev libpng-dev libtiff-dev`<br>
`sudo apt-get install libavcodec-dev libavformat-dev libswscale-dev libv4l-dev`<br>
`sudo apt-get install libxvidcore-dev libx264-dev`<br>
`sudo apt-get install libgtk-3-dev`<br>
`sudo apt-get install libcanberra-gtk*`<br>
`sudo apt-get install libatlas-base-dev gfortran`<br>
`udo apt-get install python3-dev`<br>

## Download OpenCV
Go to your home directory (or a directory of your choice<br>
`wget -O opencv.zip https://github.com/opencv/opencv/archive/4.0.0.zip`<br>
`wget -O opencv_contrib.zip https://github.com/opencv/opencv_contrib/archive/4.0.0.zip`<br>
<br>
Unzip the archives you just downloaded<br>
`unzip opencv.zip`<br>
`unzip opencv_contrib.zip`<br>

Rename the archives to remove the version number
`mv opencv-4.0.0 opencv`
`mv opencv_contrib-4.0.0 opencv_contrib`

## Setup python environment
Setup a virtual python environment
`virtualenv *name*`

Install numpy
`pip3 install numpy`

## Build/Configure OpenCV
Ensure you are still in your virtual environment
`cd ~/opencv`
`mkdir build`
`cmake -D CMAKE_BUILD_TYPE=RELEASE \
    -D CMAKE_INSTALL_PREFIX=/usr/local \
    -D OPENCV_EXTRA_MODULES_PATH=~/opencv_contrib/modules \
    -D ENABLE_NEON=ON \
    -D ENABLE_VFPV3=ON \
    -D BUILD_TESTS=OFF \
    -D OPENCV_ENABLE_NONFREE=ON \
    -D INSTALL_PYTHON_EXAMPLES=OFF \
    -D BUILD_EXAMPLES=OFF ..`

## Increase swap size (temporarily)
`sudo nano /etc/dphys-swapfile`
Update the size to 
`CONF_SWAPSIZE=2048`

Stop the swap service
`sudo /etc/init.d/dphys-swapfile stop`

Start the swap service
`sudo /etc/init.d/dphys-swapfile start`

## Install OpenCV

`make -j4`
`sudo make install`
`sudo ldconfig`

Make sure to decrease the swap size back to 100

## Link OpenCV

Find the correct cv2.cpython-35m-arm-linux-gnueabihf.so
sudo find / -name 'cv2.cpython-35m-arm-linux-gnueabihf.so'
cd ~/.virtualenvs/your_virtual_env_here/lib/python3.5/site-packages/
ln -s /location_of_module/cv2.cpython-35m-arm-linux-gnueabihf.so cv2.so
cd ~

```python
# import the necessary packages
from picamera.array import PiRGBArray
from picamera import PiCamera
import time
import cv2
 
# initialize the camera and grab a reference to the raw camera capture
camera = PiCamera()
camera.resolution = (640, 480)
camera.framerate = 32
rawCapture = PiRGBArray(camera, size=(640, 480))
 
# allow the camera to warmup
time.sleep(0.1)
 
# capture frames from the camera
for frame in camera.capture_continuous(rawCapture, format="bgr", use_video_port=True):
    # grab the raw NumPy array representing the image, then initialize the timestamp
    # and occupied/unoccupied text
    image = frame.array
 
    # show the frame
    cv2.imshow("Frame", image)
    key = cv2.waitKey(1) & 0xFF
 
    # clear the stream in preparation for the next frame
    rawCapture.truncate(0)
 
    # if the `q` key was pressed, break from the loop
    if key == ord("q"):
        break
```
